```python
# In practice, use statsmodels for VAR
from statsmodels.tsa.api import VAR

model = VAR(data[['oil_returns', 'inr_returns']])
results = model.fit(maxlags=5, ic='aic')
print(results.summary())

# Granger causality test
results.test_causality('inr_returns', causing='oil_returns', kind='f')

# Impulse response
irf = results.irf(20)
irf.plot()
```
