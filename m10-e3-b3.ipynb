```python
import torch
import torch.nn as nn

class GRUForecaster(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2, dropout=0.1):
        super().__init__()
        self.gru = nn.GRU(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            dropout=dropout,
            batch_first=True
        )
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

# Compare parameter counts
lstm_model = LSTMForecaster(input_size=1, hidden_size=64, num_layers=2)
gru_model = GRUForecaster(input_size=1, hidden_size=64, num_layers=2)

lstm_params = sum(p.numel() for p in lstm_model.parameters())
gru_params = sum(p.numel() for p in gru_model.parameters())

print(f"LSTM parameters: {lstm_params:,}")
print(f"GRU parameters:  {gru_params:,}")
print(f"GRU is {100 * (1 - gru_params/lstm_params):.0f}% smaller")
print()
print("Rule of thumb: start with GRU. Switch to LSTM only if")
print("you see the GRU underperforming on very long sequences.")
```
