```python
# In production, use statsmodels — it optimizes α, β, γ automatically
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(
    y,
    trend='add',
    seasonal='add',
    seasonal_periods=12
).fit(optimized=True)

print(f"Optimized α={model.params['smoothing_level']:.3f}")
print(f"Optimized β={model.params['smoothing_trend']:.3f}")
print(f"Optimized γ={model.params['smoothing_seasonal']:.3f}")

# Forecast next 12 months
forecast = model.forecast(12)
```
