```python
# SARIMA with statsmodels
from statsmodels.tsa.statespace.sarimax import SARIMAX

# SARIMA(1,1,1)(1,1,1,12) -- common starting point
model = SARIMAX(y,
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 12),
                enforce_stationarity=False,
                enforce_invertibility=False)
result = model.fit(disp=False)
print(result.summary())

# Forecast 24 months
forecast = result.get_forecast(steps=24)
mean_fc = forecast.predicted_mean
ci = forecast.conf_int()

plt.figure(figsize=(10, 4))
plt.plot(y, label='Observed', color='#0891B2')
plt.plot(range(n, n+24), mean_fc, label='Forecast', color='#EF4444')
plt.fill_between(range(n, n+24), ci.iloc[:,0], ci.iloc[:,1], alpha=0.2, color='#EF4444')
plt.legend()
plt.title('SARIMA Forecast: Next 24 Months')
plt.show()
```
